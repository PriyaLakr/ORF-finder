```python
def count(file_name):
    file = open(file_name)
    file.seek(0)
    records = 0
    for line in file:
        if line[0] == '>':
            records = records + 1
        else:
            continue
            
    file.close()
    print(records)
```

```python
def ORFcount(file_name):
    
    from Bio import SeqIO
    file = open(file_name)
    file.seek(0)
    choose = int(input(f" What is the minimum ORF length? ")) 
    
    for record in SeqIO.parse(file_name, "fasta"):  #it will iterate over file and print all headers and sequences
        inp = input(f" Do you want to print header and sequence? y or n ")
        if inp == "y":
            print(record.id)
            print(f"{record.seq}\n")
    
        stops = ["TAA", "TAG", "TGA"] #stop codons
        
    # Only for forward strand
    # First reading frame   
        position1 = 0     
        startcodon1 = []
        stopcodon1 = [] 
        for frame in record.seq[::]: 
            if 'ATG' and 'TAA' in record.seq[::] or 'ATG' and 'TAG' in record.seq[::] or 'ATG' and 'TGA' in record.seq[::]:
                while position1 < len(record.seq[::]) - 2:
                    codon = record.seq[::][position1:(position1+3)]  #specify codon

                    if codon == 'ATG':
                        startcodon1.append(position1)          
                    if codon in stops:
                        stopcodon1.append(position1)

                    position1 += 3    #because codons are of length 3          
            else:
                print("No ORF detected")
                
        empty = []
        stop = [0]
        ORFs = 0
        lists = []
        for i in startcodon1:
            for j in stopcodon1:
                if i < j and i not in empty and i > stop[-1]:
                    ORFseq1 = record.seq[::][i:j+3]
                    empty.append(i)
                    stop.append(j)

                    print(f"offsets: {i+1},{j+3}")
                    if len(ORFseq1) > choose:  #choose == input given by user specifying min length of ORF
                        lists.append(len(ORFseq1))
                        print(f"1st frame : {ORFseq1} and of {len(ORFseq1)} length\n")
                        ORFs += 1
                    else:
                        break 

        
        print(f"\n1st frame ORFs are {ORFs} and {lists}\n")
        
        
 # Second reading frame  
        position2 = 0
        startcodon2 = []
        stopcodon2 = []
        
        for frame in record.seq[1:]: 
            if 'ATG' and 'TAA' in record.seq[1:] or 'ATG' and 'TAG' in record.seq[1:] or 'ATG' and 'TGA' in record.seq[1:]:
                while position2 < len(record.seq[1:]) - 2:
                    codon = record.seq[1:][position2:(position2+3)]  

                    if codon == 'ATG':
                        startcodon2.append(position2)          
                    if codon in stops:
                        stopcodon2.append(position2)

                    position2 += 3              
            else:
                print("No ORF detected")
                

        empty2 = []
        stop2 = [0]
        ORFs2 = 0
        lists = []
        for i in startcodon2:
            for j in stopcodon2:
                if i < j and i not in empty2 and i > stop2[-1]:
                    ORFseq2 = record.seq[1:][i:j+3]
                    empty2.append(i)
                    stop2.append(j)
                    
                    print(f"offsets: {i+2},{j+4}")
                    if len(ORFseq2) > choose:
                        lists.append(len(ORFseq2))
                        print(f"2 frame : {ORFseq2} and of {len(ORFseq2)} length\n")
                        ORFs2 += 1
                    else:
                        break 

        
        print(f"\n2nd frame ORFs are {ORFs2} and {lists}\n")
        
# Third reading frame    
        position3 = 0
        startcodon3 = []
        stopcodon3 = []
        
        for frame in record.seq[2:]: 
            if 'ATG' and 'TAA' in record.seq[2:] or 'ATG' and 'TAG' in record.seq[2:] or 'ATG' and 'TGA' in record.seq[2:]:
                while position3 < len(record.seq[2:]) - 2:
                    codon = record.seq[2:][position3:(position3+3)]  

                    if codon == 'ATG':
                        startcodon3.append(position3)          
                    if codon in stops:
                        stopcodon3.append(position3)

                    position3 += 3              
            else:
                print("No ORF detected")
                

        empty3 = []
        stop3 = [0]
        ORFs3 = 0
        lists=[]
        for i in startcodon3:
            for j in stopcodon3:
                if i < j and i not in empty and i > stop3[-1]:
                    ORFseq3 = record.seq[2:][i:j+3]
                    empty3.append(i)
                    stop3.append(j)
                    
                    print(f"offsets: {i+3},{j+5}")
                    if len(ORFseq3) > choose:
                        lists.append(len(ORFseq3))
                        print(f"3 frame : {ORFseq3} and {len(ORFseq3)}\n")
                        ORFs3 += 1
                    else:
                        break 

        
        print(f"\n3rd frame ORFs are {ORFs3} and {lists}\n")

    file.close()

```
